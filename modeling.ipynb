# Import Packages

```python
import pandas as pd
import numpy as np
```

# Import Data

```python
#data = pd.read_csv("IICPF01.csv", skiprows=100000, nrows=8307, engine='python')
data = pd.read_csv("IICPF01.csv", engine='python')
data.shape
```

# Set Headings

```python
headers = pd.read_csv("IICPF01_desc.csv", header=None)
headers['Title'] = np.where(headers[1]=='UNUSED',headers[0],headers[1])
data.columns=headers.iloc[:,2]
data.head()
```

# Missing Data

```python
data.apply(lambda x: sum(x.isnull()))
```

# Drop Null Columns

```python
data=data.dropna(axis=1, how='all')
data.apply(lambda x: sum(x.isnull()))
```

# Convert Labels to Integers

```python
from sklearn.preprocessing import LabelEncoder

categorical_columns = [x for x in data.dtypes.index if data.dtypes[x]=='object']

le = LabelEncoder()

for col in categorical_columns:
    data[col]=data[col].fillna(' ')
    data[col] = le.fit_transform(data[col])
#    print('\nFrequency of Categories for varible %s'%col)
#    print(data[col].value_counts())
```

# Remove nulls from numeric data

```python
data=data.fillna(0)
data.apply(lambda x: sum(x.isnull()))
```

# Remove single value columns, and then:
# How many unique values in remaining 149 columns?

```python
data = data[[c for c
        in list(data)
        if len(data[c].unique()) > 1]]
data.apply(lambda x: len(x.unique()))
```

```python
data.describe()
```

# Save data for future

```python
data.to_csv("data_modified.csv",index=False)
```

# Split data into dependent / independent variables

```python
data_Y=data['CURRENT SELLING PRICE']
data_X=data.drop(['CURRENT SELLING PRICE'],axis=1)
```

# Split data into trainint / test sets

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, precision_score, recall_score, confusion_matrix, precision_recall_curve, roc_auc_score

X_train, X_test, y_train, y_test = train_test_split(data_X, data_Y, test_size=0.20, random_state=123)
```

# Run Linear Regression model

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr = lr.fit(X_train, y_train)

print ("\n\n ---Linear Regression Model---")
mse = np.mean((y_test-lr.predict(X_test))**2)

print ("Linear Regression MSE = %2.2f" % mse)
```

# Run Decision Tree model

```python
from sklearn.tree import DecisionTreeRegressor

# Decision Tree Model
dt = DecisionTreeRegressor(max_depth=15)
dt = dt.fit(X_train, y_train)

print ("\n\n ---Decision Tree Regressor---")
mse = np.mean((y_test-dt.predict(X_test))**2)

print ("Decision Tree MSE = %2.2f" % mse)
```

# Create Random Forest model

```python
from sklearn.ensemble import RandomForestRegressor

# Random Forest Model
rf = RandomForestRegressor()
rf = rf.fit(X_train, y_train)

print ("\n\n ---RandomForestRegressor---")
mse = np.mean((y_test-rf.predict(X_test))**2)

print ("Random Forest MSE = %2.2f" % mse)
```

# Show sample output, prediction vs. actual

```python
results=pd.DataFrame(np.concatenate((y_test.values[:,None],dt.predict(X_test)[:,None]),axis=1))
results['diff']=results[0]-results[1]
results.head()
```
